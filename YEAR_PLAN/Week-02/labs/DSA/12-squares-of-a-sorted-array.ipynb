# Problem 12 — Squares of a Sorted Array

- **LeetCode**: [#977](https://leetcode.com/problems/squares-of-a-sorted-array/)
- **Difficulty**: Easy (but the O(n) trick is non-obvious)
- **Pattern**: Two Pointers — *converging from both ends*, fill output *from the back*

## Problem (in my own words)
Given an array `nums` sorted in **non-decreasing** order (can include negatives), return a new array of the squares of each number, also sorted non-decreasingly.

## Naive approach
Square everything and sort: **O(n log n)**. Works but wastes the input's sortedness.

## Why two-pointers works — the key insight
After squaring, the **largest** value can only come from either end of the input (most-negative or most-positive original value). The smallest values sit near zero, wherever zero sits in the sorted array. So:
- Compare `|nums[left]|` and `|nums[right]|`, pick the larger, write it at the **back** of the result.
- Shrink that side's pointer.
- Repeat until pointers cross.

Result fills back-to-front in descending-then-descending order = final ascending order.

```python
class Solution:
    def sortedSquares(self, nums):
        n = len(nums)
        result = [0] * n
        left = 0
        right = n - 1
        write_idx = n - 1  # fill result from the back (largest squares go first)

        while left <= right:
            if abs(nums[left]) > abs(nums[right]):
                result[write_idx] = nums[left] ** 2
                left += 1
            else:
                result[write_idx] = nums[right] ** 2
                right -= 1
            write_idx -= 1

        return result
```

```python
def test(nums, expected):
    got = Solution().sortedSquares(nums)
    assert got == expected, f"mismatch: got {got}, expected {expected}"
    print(f"OK  {nums} -> {got}")

test([-4, -1, 0, 3, 10],     [0, 1, 9, 16, 100])
test([-7, -3, 2, 3, 11],     [4, 9, 9, 49, 121])
test([1, 2, 3, 4],           [1, 4, 9, 16])         # all positive
test([-5, -4, -3, -2, -1],   [1, 4, 9, 16, 25])     # all negative
test([0],                    [0])                    # single element
test([-1, 1],                [1, 1])                 # duplicates after squaring
test([-10000, 10000],        [100000000, 100000000]) # boundary values
```

## Pattern & Complexity

| | |
|---|---|
| **Pattern** | Two Pointers — converging from both ends + write-from-back |
| **Time**    | O(n) — each element is compared/written exactly once |
| **Space**   | O(n) — for the output array (unavoidable, we can't overwrite input since magnitudes shift) |

## One-line pattern note
**Two Pointers (converging): when the input is sorted but a transformation reorders magnitudes, extremes come from the ends. Fill output from the back to preserve order without sorting.**

## Why write-from-back matters
If you tried to fill the result from `result[0]`, you'd need the *smallest* square first — but small squares sit near zero, in the middle, and two pointers coming from the ends can't cheaply find them. Writing from the back lets you exploit that the *largest* square is always at one of the ends.
