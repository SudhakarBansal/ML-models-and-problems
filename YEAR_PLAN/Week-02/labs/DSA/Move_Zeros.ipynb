```python
class Solution:
    def moveZeroes(self, nums):
        """
        Do not return anything, modify nums in-place instead.
        """
        j = 0
        for i in range(len(nums)):
            if (nums[i]==0):
                continue
            nums[i], nums[j] = nums[j], nums[i]
            j+=1
        return
            
ob = Solution()
nums = [0,1,0,3,12]
nums = [0]
nums = [0,1,0,0,0,3,12]
nums = [0,1,3,12]


ob.moveZeroes(nums)
print(nums)
        
```
