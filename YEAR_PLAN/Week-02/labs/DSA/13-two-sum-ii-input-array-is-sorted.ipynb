# Problem 13 — Two Sum II (Input Array Is Sorted)

- **LeetCode**: [#167](https://leetcode.com/problems/two-sum-ii-input-array-is-sorted/)
- **Difficulty**: Medium (first Two Pointers medium in my progression)
- **Pattern**: Two Pointers — *converging from both ends*

## Problem (in my own words)
Given a 1-indexed array `numbers` sorted in non-decreasing order and a `target`, find two numbers that sum to `target`. Return their 1-based indices. The input guarantees exactly one solution and we can't use the same element twice.

## Contrast with Problem 1 — Two Sum (unsorted)
- Classic Two Sum (unsorted): hashmap of value → index, O(n) time, **O(n) space**.
- Two Sum II (sorted): two pointers exploit sortedness, O(n) time, **O(1) space**.

The sorted version lets us drop the hashmap. That's the whole point of the problem.

```python
class Solution:
    def twoSum(self, numbers, target):
        left = 0
        right = len(numbers) - 1

        while left < right:
            current_sum = numbers[left] + numbers[right]
            if current_sum > target:
                # sum too big -> shrink from the larger end
                right -= 1
            elif current_sum < target:
                # sum too small -> grow from the smaller end
                left += 1
            else:
                # LeetCode expects 1-indexed positions
                return [left + 1, right + 1]
```

```python
def test(numbers, target, expected):
    got = Solution().twoSum(numbers, target)
    assert got == expected, f"mismatch: got {got}, expected {expected}"
    print(f"OK  numbers={numbers}  target={target}  ->  {got}")

test([2, 7, 11, 15],           9,   [1, 2])
test([2, 3, 4],                6,   [1, 3])
test([-1, 0],                 -1,   [1, 2])  # negatives
test([-1, -1, 1, 1, 1],       -2,   [1, 2])  # duplicates + negatives
test([2, 3, 4, 5, 7, 9, 11, 15], 10, [2, 5]) # larger input
test([1, 2, 3, 4, 4, 9, 56, 90], 8,  [4, 5]) # duplicates at answer
```

## Pattern & Complexity

| | |
|---|---|
| **Pattern** | Two Pointers — converging from both ends, decision rule on sum vs target |
| **Time**    | O(n) — each pointer moves monotonically; together they traverse at most `n` positions |
| **Space**   | O(1) — no extra data structure |

## One-line pattern note
**Two Pointers (converging on sorted array): comparison against target dictates which side moves. Bigger → shrink right; smaller → grow left.**

## Why this works — correctness argument
At any step, the current `(left, right)` pair bounds the search space. If `numbers[left] + numbers[right] > target`, then pairing `numbers[right]` with anything `>= numbers[left]` is also too large → we can safely discard `right`. Symmetric reasoning for the `< target` case. So each move eliminates one element from the search space without missing the answer.

## Gotchas I hit
- LeetCode expects **1-indexed** return (common source of silent WAs).
- Loop condition is `left < right` (strict) — same-index pairs aren't allowed.
