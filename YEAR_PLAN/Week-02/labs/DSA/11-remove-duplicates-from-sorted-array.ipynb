# Problem 11 — Remove Duplicates from Sorted Array

- **LeetCode**: [#26](https://leetcode.com/problems/remove-duplicates-from-sorted-array/)
- **Difficulty**: Easy
- **Pattern**: Two Pointers — *slow/fast* (in-place overwrite)

## Problem (in my own words)
Given a sorted array `nums`, remove duplicates in-place so each unique element appears once. Return the length `k` of the unique prefix — the first `k` slots of `nums` must hold those unique values in their original order. Anything past index `k` is scratch.

## Key observation
Because the input is **sorted**, all duplicates are adjacent. So whenever the scanner points at a value different from the last-written unique value, we've found the next unique one.

```python
class Solution:
    def removeDuplicates(self, nums):
        # write_idx: position of the last unique value we've written
        # scan_idx:  walks forward looking for the next new value
        write_idx = 0
        scan_idx = 0
        while scan_idx < len(nums):
            if nums[scan_idx] != nums[write_idx]:
                nums[write_idx + 1] = nums[scan_idx]
                write_idx += 1
            scan_idx += 1
        return write_idx + 1
```

```python
def test(nums, expected_k, expected_prefix):
    k = Solution().removeDuplicates(nums)
    assert k == expected_k, f"k mismatch: got {k}, expected {expected_k}"
    assert nums[:k] == expected_prefix, f"prefix mismatch: got {nums[:k]}, expected {expected_prefix}"
    print(f"OK  k={k}  prefix={nums[:k]}")

test([1, 1, 2],                           2, [1, 2])
test([0, 0, 1, 1, 1, 2, 2, 3, 3, 4],      5, [0, 1, 2, 3, 4])
test([1],                                 1, [1])           # single element
test([1, 1, 1, 1],                        1, [1])           # all duplicates
test([1, 2, 3, 4, 5],                     5, [1, 2, 3, 4, 5])  # no duplicates
test([-3, -3, -1, 0, 0, 2],               4, [-3, -1, 0, 2])   # negatives included
```

## Pattern & Complexity

| | |
|---|---|
| **Pattern** | Two Pointers — slow/fast variant |
| **Time**    | O(n) — single pass |
| **Space**   | O(1) — in-place, no extra buffer |

## One-line pattern note
**Two Pointers (slow/fast): slow tracks where to *write*, fast scans for new values. Works because input is sorted so duplicates are adjacent.**

## Gotchas I hit
- Off-by-one on the return: it's `write_idx + 1` (count, not index).
- Initial `scan_idx = 0` with `write_idx = 0` is safe — the first comparison `nums[0] != nums[0]` is false, so we just advance `scan_idx` and move on. (A cleaner version starts `scan_idx = 1`.)
