```python
class Solution:
    def maxSubArray(self, nums):
        max_sub = nums[0]
        curr_sum = nums[0]

        for i in range(1, len(nums)):
            print(curr_sum)
            curr_sum = max(nums[i], curr_sum + nums[i])
            max_sub = max(max_sub, curr_sum)

        return max_sub
        
ob = Solution()
nums = [-2,1,-3,4,-1,2,1,-5,4]
# nums = [-3, -2, -5]
print(ob.maxSubArray(nums))
```
