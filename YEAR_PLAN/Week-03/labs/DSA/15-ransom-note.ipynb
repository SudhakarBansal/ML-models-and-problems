```python
class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        hash_list = [0]*26
        
        for c in magazine:
            hash_list[ord(c) - ord('a')] +=1
        
        for c in ransomNote:
            i = ord(c) - ord('a')
            if(hash_list[i] == 0):
                return False
            hash_list[i]-=1
        return True
                
ob = Solution()

ransomNote = "a"
magazine = "b"

# ransomNote = "aa"
# magazine = "ab"

# ransomNote = "aa"
# magazine = "aab"

print(ob.canConstruct(ransomNote,magazine))
```

#### pattern - frequence count of magzine -> consume elements based on ransome note -> false if count is 0 for a character. 

time complexity = O(m+n) m = size of magzine, n for size of ransomenote
space complextity = O(1) because the size of the array is fixed and other vairables are also fixed
