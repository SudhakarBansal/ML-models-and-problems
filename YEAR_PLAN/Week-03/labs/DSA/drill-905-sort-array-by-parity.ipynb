## Problem
Given an integer array `nums`, rearrange it so every **even** element appears before every **odd** element. Return the array. **Order within evens or within odds doesn't matter.**

**Examples**
- `[3,1,2,4]` → `[2,4,3,1]` (or `[4,2,3,1]`, or `[2,4,1,3]` — any valid arrangement works)
- `[0]`       → `[0]`

**Constraints**: `1 <= len(nums) <= 5000`, `0 <= nums[i] <= 5000`.

```python
class Solution:
    def sortArrayByParity(self, nums):
        # slow = next even slot; fast scans for evens; swap when fast finds an even.
        left = 0
        for right in range(len(nums)):
            if(nums[right] % 2 == 0):
                nums[left],nums[right] = nums[right],nums[left]
                left+=1
        return nums
```

```python
def is_valid(original, result):
    # answer order isn't fixed — verify multiset match + all evens before all odds
    if sorted(original) != sorted(result):
        return False, 'multiset mismatch'
    seen_odd = False
    for x in result:
        if x % 2 == 1:
            seen_odd = True
        elif seen_odd:
            return False, f'even {x} appears after an odd'
    return True, 'ok'

def test(nums):
    got = Solution().sortArrayByParity(nums.copy())
    ok, msg = is_valid(nums, got)
    assert ok, f'fail on {nums}: got {got} — {msg}'
    print(f'OK  {nums}  ->  {got}')

test([3,1,2,4])
test([0])
test([])                  # empty
test([2,4,6,8])           # all even
test([1,3,5,7])           # all odd
test([1,2])               # one of each, odd first
test([2,1])               # one of each, even first
test([0,1,0,1,0,1])       # alternating
test([5,4,3,2,1])         # reverse
```
