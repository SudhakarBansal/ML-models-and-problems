# Problem 16 — Container With Most Water

- **LeetCode**: [#11](https://leetcode.com/problems/container-with-most-water/)
- **Difficulty**: Medium
- **Pattern**: Two Pointers — Converging (decision rule: move the **shorter wall**)
- **Target time**: 25-35 min before hint

## Problem (in your own words)
Given `height = [h_0, h_1, ..., h_{n-1}]`, choose two indices `i < j` to form a container. The water it holds is `min(h_i, h_j) * (j - i)`. Return the **maximum** water any pair can hold.

**Examples**
- `[1,8,6,2,5,4,8,3,7]` → `49`  (pair `i=1, j=8` → `min(8,7)*7 = 49`)
- `[1,1]`              → `1`

**Constraints**: `2 <= n <= 10^5`, `0 <= height[i] <= 10^4`.

## Contrast with Two Sum II (Problem 13)
Two Sum II: input was sorted; decision rule was on the **sum vs target**. Here input is *not* sorted, and the decision rule is geometric — the area is gated by the **shorter** wall, so moving the taller one can never increase the min. Always move the shorter wall inward.

**Why is moving the shorter wall safe?** If we instead moved the taller wall, the new width is smaller AND the height is still capped by the (still-shorter) wall — strictly worse. So discarding the shorter wall in this step is provably non-optimal-losing.

```python
class Solution:
    def maxArea(self, height):
        # left=0, right=n-1; track best; move the shorter wall.
        left = 0
        right = len(height) - 1
        max_area = 0
        
        while left < right:
            h = min(height[left],height[right])
            l = right - left
            area = h*l
            max_area = max(max_area,area)
            
            if(height[left]>height[right]):
                right-=1
            else:
                left+=1
            
        return max_area
```

```python
def test(height, expected):
    got = Solution().maxArea(height)
    assert got == expected, f"mismatch: got {got}, expected {expected}"
    print(f"OK  height={height}  ->  {got}")

test([1,8,6,2,5,4,8,3,7], 49)
test([1,1],               1)
test([4,3,2,1,4],         16)   # equal walls at both ends — full width
test([1,2,1],             2)    # narrow valley
test([2,3,4,5,18,17,6],   17)   # tall spike near the end
test([0,2],               0)    # zero wall
```

## Pattern & Complexity

| | |
|---|---|
| **Pattern** | Two Pointers — converging; move the shorter wall |
| **Time**    | O(n) — each pointer moves monotonically inward |
| **Space**   | O(1) |
