## Problem
Given an integer array `nums`, return an array `result` of the same length where `result[i] = nums[0] + nums[1] + ... + nums[i]`. That is, each output cell is the running (prefix) sum up to and including position `i`.

**Examples**
- `[1,2,3,4]` → `[1,3,6,10]`
- `[1,1,1,1,1]` → `[1,2,3,4,5]`
- `[3,1,2,10,1]` → `[3,4,6,16,17]`

**Constraints**: `1 <= len(nums) <= 1000`, `-10^6 <= nums[i] <= 10^6`.

## The pattern (recall before coding)
**Running sum**: walk left to right, keep a single accumulator `s`, write `s` at each position. The whole point of the prefix-sum pattern is that **each cell only depends on the previous cell + the current input** — so you do one pass, O(n) time, O(1) extra space (or O(n) if you allocate the result array, which is unavoidable here).

Two valid styles:
1. **In-place**: mutate `nums` itself (`nums[i] += nums[i-1]` for `i >= 1`). Returns `nums`.
2. **Fresh array**: allocate `result` of length n, accumulate into a separate variable.

Both are O(n) time. In-place uses O(1) extra; fresh array uses O(n) extra. Either is acceptable.

```python
class Solution:
    def runningSum(self, nums):
        s = 0
        
        for i in range(len(nums)):
            s+= nums[i]
            nums[i] = s
        return nums
```

```python
def test(nums, expected):
    got = Solution().runningSum(nums.copy())  # .copy() so in-place mutation doesn't pollute later tests
    assert got == expected, f'fail on {nums}: got {got}, expected {expected}'
    print(f'OK  {nums}  ->  {got}')

test([1,2,3,4],          [1,3,6,10])
test([1,1,1,1,1],        [1,2,3,4,5])
test([3,1,2,10,1],       [3,4,6,16,17])
test([5],                [5])              # single element
test([0,0,0,0],          [0,0,0,0])        # all zeros
test([-1,-2,-3,-4],      [-1,-3,-6,-10])   # all negatives
test([10,-5,3,-2],       [10,5,8,6])       # mixed signs (the running sum can decrease)
test([100,-50,-50,100],  [100,50,0,100])   # cancellation back to 0
```
