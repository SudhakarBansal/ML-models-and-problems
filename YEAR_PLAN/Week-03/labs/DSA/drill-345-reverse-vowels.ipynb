## Problem (in your own words — fill before coding)
Given a string `s`, reverse only the vowels (`a, e, i, o, u`, both cases) and return the result. Non-vowels stay in place.

**Examples**
- `"IceCreAm"` → `"AceCreIm"`
- `"leetcode"` → `"leotcede"`
- `"hello"`    → `"holle"`

**Constraints**: `1 <= len(s) <= 3 * 10^5`. `s` consists of printable ASCII characters.

```python
class Solution:
    def reverseVowels(self, s: str) -> str:
        # converging two pointers; swap when both sides are vowels
        
        vowels = set('aeiouAEIOU')
        s = list(s)
        i = 0
        j = len(s) - 1
        while i < j:
            if s[i] not in vowels:
                i += 1
            elif s[j] not in vowels:
                j -= 1
            else:
                s[i], s[j] = s[j], s[i]
                i += 1
                j -= 1
        return ''.join(s)
```

```python
def test(s, expected):
    got = Solution().reverseVowels(s)
    assert got == expected, f"mismatch: got {got!r}, expected {expected!r}"
    print(f"OK  {s!r}  ->  {got!r}")

test("IceCreAm", "AceCreIm")
test("leetcode", "leotcede")
test("hello",    "holle")
test("a",        "a")            # single vowel
test("bcd",      "bcd")          # no vowels
test("AEIOU",    "UOIEA")        # all upper-case vowels
test("aA",       "Aa")           # case preserved at swap site
```
