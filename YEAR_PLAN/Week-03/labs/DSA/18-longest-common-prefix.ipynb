# Problem 18 — Longest Common Prefix

- **LeetCode**: [#14](https://leetcode.com/problems/longest-common-prefix/)
- **Difficulty**: Easy
- **Pattern**: Strings — vertical scan (column-wise comparison)
- **Target time**: 15-20 min — quick win

## Problem (in your own words)
Given a list of strings `strs`, return the longest prefix shared by **all** of them. If no common prefix exists, return `""`.

**Examples**
- `["flower","flow","flight"]` → `"fl"`
- `["dog","racecar","car"]`    → `""`
- `["interspecies","interstellar","interstate"]` → `"inters"`

**Constraints**: `1 <= len(strs) <= 200`, `0 <= len(strs[i]) <= 200`. Strings consist of lowercase English letters.

## Approaches to consider (pick one cold)
1. **Vertical scan** — for column index `i = 0, 1, 2, ...`, check `strs[0][i]` against every other string at index `i`. Stop when a mismatch (or any string runs out) appears. Return `strs[0][:i]`.
2. **Horizontal scan** — start with `prefix = strs[0]`; for each next string, shrink `prefix` until it's a prefix of that string.
3. **Sort + compare ends** — sort the list; the LCP of the **first** and **last** string after sort = LCP of all (because lex-extremes bound the rest).

Vertical scan is usually the cleanest first answer.

```python
class Solution:
    def longestCommonPrefix(self, strs):
        # vertical scan: walk column i; break on mismatch or string-end; return strs[0][:i].
        pref = ""        
        j = 0
        while j<len(strs[0]):
            pref=strs[0][j]
            for i in range(len(strs)):
                if(len(strs[i])<=j or strs[i][j] != pref):
                    return strs[0][0:j]
            j+=1
        return strs[0]
```

```python
def test(strs, expected):
    got = Solution().longestCommonPrefix(strs)
    assert got == expected, f"mismatch on {strs}: got {got!r}, expected {expected!r}"
    print(f"OK  {strs}  ->  {got!r}")

test(["flower","flow","flight"],                  "fl")
test(["dog","racecar","car"],                     "")
test(["interspecies","interstellar","interstate"], "inters")
test(["a"],                                       "a")          # single string
test(["","abc"],                                  "")           # one empty input
test(["abc","abc","abc"],                         "abc")        # all identical
test(["ab","a"],                                  "a")          # one is a prefix of the other
test(["c","c"],                                   "c")          # all length-1 same
```

## Pattern & Complexity

| | |
|---|---|
| **Pattern** | Vertical scan over a list of strings |
| **Time**    | O(S) where S = total chars across all strings (worst case) |
| **Space**   | O(1) extra (or O(L) for the returned prefix) |

## One-line pattern note (write from memory after solving)
_e.g._ "Walk character columns of strs[0]; stop at first column where any other string disagrees or has run out. Return prefix up to that column."

## Edge cases that bite
- Empty input list — problem guarantees ≥1, but defensive check is cheap.
- One of the strings is empty — answer is `""`.
- All strings identical — answer is the string itself.
- One string is a strict prefix of another — answer is the shorter string.
