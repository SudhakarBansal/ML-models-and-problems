# Problem 17 — Valid Parentheses  *(first Stack problem)*

- **LeetCode**: [#20](https://leetcode.com/problems/valid-parentheses/)
- **Difficulty**: Easy
- **Pattern**: **Stack** — match each closing bracket with the most recent unmatched opening bracket
- **Target time**: 20-25 min

## Problem (in your own words)
Given a string `s` containing only `(`, `)`, `{`, `}`, `[`, `]`, decide whether the brackets are **balanced and properly nested**.

Rules:
1. Every opener must be closed by the **same** type.
2. Openers must be closed in the **correct order** (innermost first).
3. Every closer must have a corresponding earlier opener.

**Examples**
- `"()"`     → `True`
- `"()[]{}"` → `True`
- `"(]"`     → `False`
- `"([)]"`   → `False`   *(nested wrong)*
- `"{[]}"`   → `True`

## Why a stack? (the pattern)
"I have to match an open with the **most recent unmatched** open." That is a textbook LIFO operation: push every opener; on a closer, pop and check the top matches. If the stack is empty when you see a closer, the closer has no opener → fail. If the stack is non-empty at the end, some openers were never closed → fail.

```python
class Solution:
    def isValid(self, s: str) -> bool:
        # push openers; on closer, pop & check match; pair table {')':'(', ']':'[', '}':'{'}.
        pairs = {')': '(', ']': '[', '}': '{'}
        stack = list()
        for c in s:
            if c in pairs:
                if not stack or stack.pop() != pairs[c]:
                    return False
            else:
                stack.append(c)
        return not stack
```

```python
def test(s, expected):
    got = Solution().isValid(s)
    assert got == expected, f"mismatch on {s!r}: got {got}, expected {expected}"
    print(f"OK  {s!r:>14}  ->  {got}")

test("()",          True)
test("()[]{}",      True)
test("(]",          False)
test("([)]",        False)
test("{[]}",        True)
test("",            True)    # empty string is trivially balanced
test("(",           False)   # never closed
test(")",           False)   # closer with no opener
test("((((",        False)
test("){",          False)   # closer first
test("({[]})",      True)    # deep nesting
```

## Pattern & Complexity

| | |
|---|---|
| **Pattern** | Stack — LIFO match for nested structures |
| **Time**    | O(n) — each char is pushed and popped at most once |
| **Space**   | O(n) — stack holds up to n openers (e.g. `"(((("`) |


## Two failure modes — don't conflate them
1. **Wrong type pop** (e.g. `"(]"` — top is `(` but closer is `]`) → pop returned a non-matching opener.
2. **Empty pop** (e.g. `")"`) → no opener at all. Many implementations crash here; handle it.
