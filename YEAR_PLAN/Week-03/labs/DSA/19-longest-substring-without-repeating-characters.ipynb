# Problem 19 — Longest Substring Without Repeating Characters  *(first variable-size Sliding Window)*

- **LeetCode**: [#3](https://leetcode.com/problems/longest-substring-without-repeating-characters/)
- **Difficulty**: Medium
- **Pattern**: **Sliding Window — variable-size** (the canonical example)
- **Target time**: 30-40 min

## Problem
Given a string `s`, return the length of the **longest substring** that contains **no repeated characters**.

**Examples**
- `"abcabcbb"` → `3`  (`"abc"`)
- `"bbbbb"`    → `1`  (`"b"`)
- `"pwwkew"`   → `3`  (`"wke"` — note: `"pwke"` is a *subsequence*, not a substring)

**Constraints**: `0 <= len(s) <= 5*10^4`. `s` has English letters, digits, symbols, spaces.

## The pattern (variable-size sliding window)

Maintain `[l, r]` as the current window. Track which characters are inside (set or hashmap of last-seen index).

```
for r in range(n):
    while window_contains(s[r]):     # invariant broken
        remove s[l]; l += 1          # shrink from left until invariant restored
    add s[r] to window
    best = max(best, r - l + 1)
```

**Why O(n)**: each character is added once and removed at most once, so each pointer moves monotonically forward. Total work is 2n, not n².

````python
class Solution:
    def lengthOfLongestSubstring(self, s):
        hash_map = dict()
        max_len = 0
        left=0
        
        for right in range(len(s)):
            if(s[right] in hash_map):
                left = max(left,hash_map[s[right]]+1)
                
            hash_map[s[right]] = right
            max_len=max(max_len,right-left+1)
        return max_len
````

````python
ob = Solution()
print(ob.lengthOfLongestSubstring('abcabcbb'))
````

````python
def test(s, expected):
    got = Solution().lengthOfLongestSubstring(s)
    assert got == expected, f'fail on {s!r}: got {got}, expected {expected}'
    print(f'OK  {s!r:>14}  ->  {got}')

test('abcabcbb', 3)
test('bbbbb',    1)
test('pwwkew',   3)
test('',         0)            # empty
test('a',        1)            # single char
test('au',       2)            # all unique, len 2
test('dvdf',     3)            # 'vdf' — left pointer must jump past first 'd'
test('abba',     2)            # 'ab' then 'ba' — careful with stale indices
test(' ',        1)            # space is a valid char
test('aab',      2)            # the middle 'a' should be skipped, then 'ab'
test('tmmzuxt',  5)            # 'mzuxt' — non-trivial reset
````

## Pattern & Complexity

| | |
|---|---|
| **Pattern** | Sliding Window — variable-size; expand right, shrink left when invariant breaks |
| **Time**    | O(n) — each char enters and leaves the window at most once |
| **Space**   | O(min(n, charset)) — ≤ 26 for lowercase, ≤ 128 for ASCII |
