```python
class Solution:
    def merge(self, nums1, m, nums2, n):
        """
        Do not return anything, modify nums1 in-place instead.
        """
        
        i=j=0
        
        while(i<m+n):
            if(nums2[j] < nums1[i]):
                nums1.insert(i,nums2[j])
                j+=1
            i+=1   
            print(nums1)
        
        return
```

```python
ob = Solution()
nums1 = [1,2,3,0,0,0]
nums2 = [2,5,6]
m = 4
n = 3
ob.merge(nums1,m,nums2,n)
print(nums1)
```

```python
class Solution:
    def productExceptSelf(self, nums):
        n = len(nums)
        pref = [1]*n
        res = [1]*n
        suff = [1]*n
        for i in range(1,n):
            pref[i] = pref[i-1] * nums[i-1]
            suff[-i-1] = suff[-i] * nums[-i]
                    
        for i in range(n):
            res[i] = pref[i] * suff[i]

        return res
    
ob = Solution()
nums = [1,2,3,4]
print(ob.productExceptSelf(nums))
```
